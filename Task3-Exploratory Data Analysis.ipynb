```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Removing unwanted columns in Dataset
df = data.drop(['Postal Code'], axis=1)
```

<font size="4">**Exploratory Data Analysis**</font>

<font size="4">**Visualizing the whole dataset**</font>

```python
import seaborn as sns
sns.pairplot(df)
```

```python
# Finding correlation between columns using heatmap
df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

<font size="4">**Visualizing the Categories:**</font>

```python

plt.figure(figsize=(6,6))
textprops={"fontsize":15}
plt.title('Category')
plt.pie(df['Category'].value_counts(),labels=df['Category'].value_counts().index,autopct='%1.1f%%',textprops=textprops)
plt.show()
```

```python
# Plotting Profit Vs Sales graph for categories
plt.figure(figsize=(10,16))
df.groupby('Category')['Profit', 'Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

<font size="2">**Observation: The Furniture category is giving is much less profit than the other two categories.**</font>

<font size="4">**Visualizing Sub-Categories:**</font>

```python
# computing sub-categories in terms of sales from top 100 observations
top_s=df.groupby("Sub-Category").Sales.sum().nlargest(n=100)

# computing sub-categories in terms of profit from top 100 observations
top_p=df.groupby("Sub-Category").Profit.sum().nlargest(n=100)

plt.style.use('seaborn')
top_s.plot(kind='bar',figsize=(10,5),fontsize=15)
top_p.plot(kind='bar',figsize=(10,5),fontsize=15,color='red')
plt.xlabel('Sub-Category',fontsize=15)
plt.ylabel('Total Sales Vs Profit',fontsize=15)
plt.title("Sub-Category Sales Vs Profit (Top 100 Observations)", fontsize=15)
plt.show()
```

<font size="2">**Observation: Tables are giving a Negative Profit i.e. Loss while Copiers and Phones are giving Maximum Profit**</font>

<font size="4">**Visualizing Sales Vs Profit in Different Regions:**</font>

```python
plt.figure(figsize=(6,6))
plt.title('Region')
plt.pie(df['Region'].value_counts(),labels=df['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

<font size="2">**Observation: The East and West regions are having more Profit than compared to Central and South region**</font>

<font size="3">**Visualizing Sales Vs Profit in different States:**</font>

```python
# Computing states in terms of profit from first 10 observations
top_states=df.groupby("State").Sales.sum().nlargest(n=10)

# Computing states in terms of profit from first 10 observations
top_profit=df.groupby("State").Profit.sum().nlargest(n=10)

plt.style.use('seaborn')
top_states.plot(kind='bar',figsize=(10,5),fontsize=15)
top_profit.plot(kind='bar',figsize=(10,5),fontsize=15,color='red')
plt.xlabel('States',fontsize=15)
plt.ylabel('Total Sales Vs Profit',fontsize=15)
plt.title("State's Sales Vs Profit(Top 10 Observations)",fontsize=15)
plt.show()
```

<font size="2">**Observation: States like California and New York are giving more profit than other states**</font>

<font size="3">**Interdependency of Sales, Profit and Discount:**</font>

```python
plt.style.use('seaborn')
df.plot(kind="scatter",figsize=(10,5),x="Sales",y="Profit",c="Discount",s=20,fontsize=15,colormap="viridis")
plt.ylabel('Total Profits',fontsize=15)
plt.title("Interdependency Of Sales, Profit And Discount",fontsize=15)
plt.show()
```

<font size="2">**Observation: If the Discount is increased, the Sales will also increase but we'll have to face the Loss**</font>
